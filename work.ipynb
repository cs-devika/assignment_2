```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
avg_petal = iris.groupby("species")["petal_length"]
plt.bar(iris["species"], iris["petal_length"])
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.title("Average Petal Length of Iris Species")
plt.show()
```

```python
petal_length = iris.groupby("species")["petal_length"].first()
petal_width  = iris.groupby("species")["petal_width"].first()
plt.bar(petal_length.index, petal_length.values, color="g")
plt.bar(petal_width.index, petal_width.values, color="b")
```

```python
plt.barh(petal_length,petal_width)
```

```python
plt.scatter(petal_length,petal_width ,color = "g")
plt.show()
```

```python
fig,ax = plt.subplots(1,3,figsize = (12,4))
ax[0].plot(petal_length,petal_width ,color="olive")
ax[1].bar(petal_length,petal_width ,color ="pink")
ax[2].scatter(petal_length,petal_width ,color="yellow")
plt.show()
```

```python
plt.pie(petal_length,labels = petal_width)
plt.show()
```
