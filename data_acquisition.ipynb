```python
import sqlite3
import requests
import pandas as pd
import numpy as np
```

```python
url = "/content/launches.json"
launches= pd.read_json(url)
launches
```

```python
launches = launches[["name", "date_utc", "success", "details", "rocket"]]
launches
```

```python
launches["date_utc"] = pd.to_datetime(launches["date_utc"])
launches["year"] = launches["date_utc"].astype(str).str.split("-").str[0]
launches[["date_utc", "year"]].head()
```

```python
url = "/content/rockets.json"
rockets= pd.read_json(url)
rockets
```

```python
rockets= rockets[["id", "name", "type", "active", "stages"]]
rockets
```

```python
launches = launches.rename(columns={'name': 'launch_name'})
rockets = rockets.rename(columns={'name': 'rocket_name'})
```

```python
merged_df = pd.merge(
    launches,
    rockets,
    left_on='rocket',
    right_on='id',
    how='left'
)
(merged_df)
```

```python
countrys = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country'] = np.random.choice(countrys, size=len(merged_df))
merged_df
```

```python
conn = sqlite3.connect('launches.db')
merged_df.to_sql('launches', conn, index=False)
```

```python
data= pd.read_sql('select * from launches', conn)
data.head()
```

```python
cursor=conn.cursor()
```

```python
cursor.execute("""
    select country,count(*) as launches_count
    from launches
    group by country
    order by launches_count desc;
    """
)
```

```python
rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
    select year,count(*) as launches_count
    from launches
    group by year
    order by launches_count desc;
""")
```

```python
row = cursor.fetchone()
print(row)
```

```python
cursor.execute("""
    select launch_name,count(*) as launches_count
    from launches
    group by launch_name
    order by launches_count desc
    limit 5;
""")
```
