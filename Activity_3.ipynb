Create a Python program to calculate the factorial of a given number using
a function.

```python
def fact(n):
  f = 1
  for i in range(1,n+1):
    f = f*i
  return f
n = int(input("Enter the number : "))
print("factorial :", fact(n))
```

Develop a Python program that checks whether a given word or phrase is a
palindrome.

```python
a = input("Enter the word or number : ")
if a == a[::-1]:
  print(a," is palindrome")
else:
  print(a," is not palindrome")
```

Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
a = input("Enter the input : ")
vowels = 0
consonants = 0
for i in a.lower():
  if i in "a,e,i,o,u":
    vowels += 1
  elif i.isalpha():
    consonants += 1
print("vowels : ",vowels)
print("consonants : ",consonants)
```

Develop a Python program that takes two lists as input and returns a new
list containing the common elements between them.

```python
a = [1,2,3,4,5,6,7]
b = [1,3,5,7]
new_list = []
for i in a:
  if i in b:
    new_list.append(i)
print(new_list)
```

Develop a Python program that checks if a given year is a leap year.

```python
num = int(input("Enter the year : "))
if (num % 400 ==0) or (num % 4 == 0 and num % 100 != 0):
  print("leap year")
else:
  print("not leap year")
```

Write a Python program that checks if a given number, is a prime number.

```python
n = int(input("Enter the number : "))
if n>1 :
  for i in range(2,n):
    if n % i == 0:
      print("it's not a prime number")
      break
  else:
    print("its prime number")
else:
  print("its not  prime")
```

Write a Python program to generate the Fibonacci sequence up to a specified number of terms.

```python
n = int(input("Enter the number : "))
a,b = 0,1
for i in range(n):
  print(a)
  a,b = b,a+b
```
