# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def range(lists):
  min = lists[0]
  # set first value of the list as min
  max = lists[0]
  # set first value of the list as max
  for i in lists:
    # looping through the entere list to check for max and min value
    if(i >= max):
      max = i
    if(i <= min):
      min = i
  return(min,max)
lists = [12,13,14,25,36,43,21]
range = range(lists)
range

```

## Q2. Write function that returns mean value of a list of numbers

```python
def list_number(num):
  sum = 0
  for i in num:
    sum += i
  mean = sum /len(num)
  return mean
num1 = (13,24,53,65,32)
value = list_number(num1)
value
```

# Q3. Write a function that returns the median value of a list of numbers

```python
def median(list_number):
  list_number = sorted(list_number)
  length = len(list_number)
  median = length//2
  if length % 2 != 0:
    return list_number[median]
list_number = [12,23,34,46,32]
median_value = median(list_number)
median_value
```
